```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Rawoo\OneDrive\Documents\Projects\walmart-sales-analysis-python\data\processed\walmart_cleaned.csv")
df.head()
```

```python
total_sales = df['Weekly_Sales'].sum()
total_sales
```

```python
avg_sales = df["Weekly_Sales"].mean()
avg_sales
```

```python
store_sales = df.groupby("Store")["Weekly_Sales"].sum().sort_values(ascending=False)
store_sales.head(10)
```

```python
store_sales.tail(10)
```

```python
yearly_sales = df.groupby("Year")["Weekly_Sales"].sum()
yearly_sales
```

```python
monthly_sales = (df.groupby(["Year", "Month","Month_Name"])["Weekly_Sales"].sum()
.reset_index().sort_values(by=["Year", "Month"]))
monthly_sales.head()
```

```python
holiday_sales = df.groupby("Holiday_Flag")["Weekly_Sales"].agg(["mean","sum","count"])
holiday_sales
```

```python
holiday_means = df.groupby("Holiday_Flag")["Weekly_Sales"].mean()
holiday_lift_pct = ((holiday_means[1] - holiday_means[0]) / holiday_means[0]) * 100
holiday_lift_pct
```
